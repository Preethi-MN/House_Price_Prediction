```python
# 6.univarite analysis
#6.1 Histogram(Distribution)
plt.figure()
plt.hist(df['Price_Lakhs'],bins = 25)
plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# 7.Bivarite Analysis
# 7.1 Scatter Plot(Area vs Price)
plt.figure()
plt.scatter(df['Area_sqft'],df['Price_Lakhs'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
plt.figure()
plt.bar(df['Area_sqft'],df['Price_Lakhs'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
# 7.2 Bar Plot(Location vs Avg Price)
df.groupby('Location')['Price_Lakhs'].mean().plot(kind='bar')
plt.title("Avg vs Location")
plt.xlabel("Location")
plt.ylabel("Average")
plt.show()
                                                
```

```python
# 8.Multivariate Analysis
# 8.1 Correlation Matrix

corr = df.corr(numeric_only = True)
print(corr)
```
