```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# training the data
X=([[1],[2],[3],[4],[5]])
y=([40,50,65,75,90])

#train the model
model=LinearRegression()
model.fit(X,y)
# Plot the data points (scatter plot)
plt.scatter(X, y, color="purple", label="Actual Marks") 

# Plot the regression line
plt.plot(X, model.predict(X), color="red", label="Regression Line")

# Labels and title 
plt.xlabel("Hours Studied") 
plt.ylabel("Marks Scored")
plt.title("Study Hours vs Marks Prediction") 
plt.legend()

# Show the plot 
plt.show()

#user input
hours=float(input("Enter how many hours you studied="))

#prediction
predicted_marks=model.predict([[hours]])[0]

#output
print(f"based on your hours {hours} you may score around {predicted_marks} ")
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
X=[[1],[2],[3],[4],[5]]
y=[0,0,1,1,1]
model=LogisticRegression()
model.fit(X,y)

#Values for smooth curve
x_plot = np.linspace(1, 5, 100).reshape(-1, 1) 
y_prob = model.predict_proba(x_plot)[:, 1]

#Graph
plt.figure() 
plt.scatter(X, y)
plt.plot(x_plot, y_prob)
plt.xlabel("Hours studied")
plt.ylabel("Probability of passing") 
plt.show() 
hours=float(input("Enter how many hours you studied="))
result=model.predict([[hours]])[0] 
if result==1: 
 print(f"based on hours {hours} you are likely to PASS")
else: 
 print(f"based on hours {hours} you are likely to FAIL")
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Training data
X = [[180, 7], [200, 7.5], [250, 8], [300, 8.5], [330, 9], [360, 9.5]]
Y = [0, 0, 0, 1, 1, 1]

# Create the classifier`1
model= KNeighborsClassifier(n_neighbors=3)

# Fit the model
model.fit(X,Y)
weight=float(input("enter the weights in gram:"))
size=float(input("enter the size in cm:"))
prediction=model.predict([[weight,size]])[0]
if prediction==0:
    print("This is  likely an Apple")
else:
    print("This is likely an Orange")

# Separate points based on class
apple_x = [X[i][0] for i in range(len(Y)) if Y[i] == 0]
apple_y = [X[i][1] for i in range(len(Y)) if Y[i] == 0]

orange_x = [X[i][0] for i in range(len(Y)) if Y[i] == 1]
orange_y = [X[i][1] for i in range(len(Y)) if Y[i] == 1]

# Plot training points
plt.figure()
plt.scatter(apple_x, apple_y, label="Apple")
plt.scatter(orange_x, orange_y, label="Orange")

# Plot user input point
plt.scatter(weight, size, marker='x', label="Test point")

plt.xlabel("Weight (grams)")
plt.ylabel("Size (cm)")
plt.legend()
plt.title("KNN Fruit Classification")

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
X=[[7,2],[8,3],[9,10],[10,9]]
Y=[0,0,1,1]
model=DecisionTreeClassifier()
model.fit(X,Y)
size=float(input("enter the fruit size in cm="))
shade=float(input("enter the color shade (1-10):"))
result=model.predict([[size,shade]])[0]
if result==0:
    print("This is likely an Apple")
else:
    print("This is likely an Orange")

import matplotlib.pyplot as plt

# Separate Apple and Orange points
apple_x = [X[i][0] for i in range(len(Y)) if Y[i] == 0]
apple_y = [X[i][1] for i in range(len(Y)) if Y[i] == 0]

orange_x = [X[i][0] for i in range(len(Y)) if Y[i] == 1]
orange_y = [X[i][1] for i in range(len(Y)) if Y[i] == 1]

# Plot
plt.figure()

plt.scatter(apple_x, apple_y, label="Apple")
plt.scatter(orange_x, orange_y, label="Orange")

# Your input point
plt.scatter(size, shade, marker='x', label="Test point")

plt.xlabel("Fruit Size (cm)")
plt.ylabel("Color Shade")
plt.title("Decision Tree – Fruit Classification")
plt.legend()

plt.show()
    
```
